```python
import pandas as pd
titanic_data = pd.read_csv("/content/sample_data/train _data/train.csv")
print(titanic_data)
```

```python
if "Cabin"  in titanic_data:
  titanic_data.drop(columns="Cabin",inplace=True)
print(titanic_data.info())
```

```python
titanic_data['age_z'] = (titanic_data['Age'] - titanic_data['Age'].mean()) / titanic_data['Age'].std()
edge_age = titanic_data[abs(titanic_data['age_z']) > 3]
print(edge_age[['Age', 'age_z']])
```

```python

titanic_data['Title'] = titanic_data['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
titanic_data['Age'] = titanic_data.groupby('Title')['Age'].transform(lambda x: x.fillna(x.median()))
print(titanic_data["Age"])
```

```python
embark= titanic_data["Embarked"].mode()
titanic_data.loc[:,"Embarked"] = titanic_data.loc[:,"Embarked"].fillna(embark[0])
print(titanic_data["Embarked"])
```

```python
print(titanic_data.info())
```

```python
titanic_data=titanic_data.drop(columns=["PassengerId"],errors='ignore')
print(titanic_data)
```

```python
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1
titanic_data['Alone'] = (titanic_data['FamilySize'] == 1).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder

input_features = titanic_data[["Title","Sex", "Age","Fare","Embarked","Pclass","FamilySize","Alone"]].copy()

le = LabelEncoder()

for col in [ "Title","Sex", "Embarked"]:
    input_features[col] = le.fit_transform(input_features[col])

print(input_features.tail())

```

```python
from sklearn.model_selection import train_test_split
X = input_features
Y = titanic_data["Survived"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
Y_train.info()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=150,
    max_depth= 6,
    min_samples_split=8,
    min_samples_leaf=7,
    max_features='sqrt',
    random_state=42
)

model.fit(X_train, Y_train)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train, Y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy:", scores.mean())
```

```python
from sklearn.metrics import accuracy_score

train_preds = model.predict(X_train)
accuracy = accuracy_score(Y_train, train_preds)
print("Training Accuracy:", accuracy)
```
